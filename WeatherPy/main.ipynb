```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
city_coords = []
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        city_coords.append(lat_lng)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#lists to be populated w/ weather data
city_name = []
country = []
lattitude =[]
longitude = []
date = []
temperature = []
wind_speed = []
cloud_cover = []
humidity = []

print(api_key)
```

```python
#creating url for api calls, setting up for loop w/ try/except to pull data from json and put it in lists

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key

counter = 1

for city in cities:  
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        
        city_name.append(response["name"])
        country.append(response["sys"]["country"])
        lattitude.append(response["coord"]["lat"])
        longitude.append(response["coord"]["lon"])
        date.append(response["dt"])
        temperature.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        cloud_cover.append(response["clouds"]["all"])
        humidity.append(response["main"]["humidity"])
        
        print(f"city  {counter} | {city}")
        print(f"{url}&q={city}")
        print(response)
        
        counter = counter + 1
        
    except:
        
        time.sleep(1.05)
        print("No city here. Moving on...")
        
    continue
print("All Done!")
```

```python
#putting my lists into a data frame to later be exported to a csv
weather_dict = {"City":city_name, 
            "Country":country, 
            "Lattitude":lattitude, 
            "Longitude":longitude,
            "Date":date,
            "Temperature":temperature, 
            "Wind Speed":wind_speed, 
            "Cloud Cover":cloud_cover,
            "Humidity":humidity}

weather_df = pd.DataFrame.from_dict(weather_dict)
weather_df.head()
```

```python
#lattitude temp graph
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
ax.yaxis.grid(color='gray')
ax.xaxis.grid(color='gray')

ax = plt.scatter(x=lattitude, y=temperature, color='green', s=20)
plt.title('City Lattitude vs High Temperature 7-27-19')
plt.ylabel('Temperature')
plt.xlabel('Lattitude')

# Save the Figure
plt.savefig('temperature_lattitude.png')

# Show the Figure
plt.show()
  
```

```python
#lattitude humidity graph
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
ax.yaxis.grid(color='gray')
ax.xaxis.grid(color='gray')

ax = plt.scatter(x=lattitude, y=humidity, color='magenta', s=20)
plt.title('City Lattitude vs Humidity 7-27-19')
plt.ylabel('Humidity')
plt.xlabel('Lattitude')

# Save the Figure
plt.savefig('humidity_lattitude.png')

# Show the Figure
plt.show()
```

```python
#lattitude cloud graph
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
ax.yaxis.grid(color='gray')
ax.xaxis.grid(color='gray')

ax = plt.scatter(x=lattitude, y=cloud_cover, color='teal', s=20)
plt.title('City Lattitude vs Cloud Cover 7-27-19')
plt.ylabel('Cloud Cover')
plt.xlabel('Lattitude')

# Save the Figure
plt.savefig('cloud_cover_lattitude.png')

# Show the Figure
plt.show()
```

```python
#lattitude wind graph
fig, ax = plt.subplots()
fig.set_size_inches(5,5)
ax.yaxis.grid(color='gray')
ax.xaxis.grid(color='gray')

ax = plt.scatter(x=lattitude, y=wind_speed, color='orange', s=20)
plt.title('City Lattitude vs Wind Speed 7-27-19')
plt.ylabel('Wind Speed')
plt.xlabel('Lattitude')

# Save the Figure
plt.savefig('wind_speed_lattitude.png')

# Show the Figure
plt.show()
```
